Linestyle ----- Dotted line

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([2, 4, 3, 7])
y = np.array([3, 6, 2, 11])
plt.plot(x, y, linestyle = "dotted")
plt.show()
```

Use a dashed line

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([2, 4, 3, 7])
y = np.array([3, 6, 2, 11])
plt.plot(x, y, linestyle = "dashed")
plt.show()
```

Line Color

```python
import matplotlib.pyplot as plt
import numpy as np
y = np.array([3, 6, 2, 11])
plt.plot(x, y, linestyle = "dashed", color = 'b')
plt.show()
```

Hexadecimal color values

```python
import matplotlib.pyplot as plt
import numpy as np
y = np.array([3, 6, 2, 11])
plt.plot(x, y, linestyle = "dashed", color = '#4CAF50')
plt.show()
```

Line Width

```python
import matplotlib.pyplot as plt
import numpy as np
y = np.array([3, 6, 2, 11, 1 , 10])
plt.plot(y, color = '#4CAF50', linewidth =20)
plt.show()
```

Draw two line

```python
import matplotlib.pyplot as plt
import numpy as np
y1 = np.array([2, 4, 3, 7])
y2 = np.array([3, 6, 2, 11])
plt.plot(y1)
plt.plot(y2)
plt.show()
```

Draw two lines by specifiyng the x- and y-point values for both lines

```python
import matplotlib.pyplot as plt
import numpy as np

x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])
x2 = np.array([0, 1, 2, 3])
y2 = np.array([6, 2, 7, 11])

plt.plot(x1, y1, x2, y2)
plt.show()
```
