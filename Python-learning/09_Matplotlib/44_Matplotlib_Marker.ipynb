Markers

Make each point with a circle

```python
import matplotlib.pyplot as plt
import numpy as np
y = np.array([3, 8, 1, 10])
plt.plot( y, marker = 'o')
plt.show()
```

Mark each point with a star

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([3, 6, 2, 5])
y = np.array([4, 5, 10, 13])
plt.plot(x, y, marker = '*')
plt.show()
```

Format Strings fmt

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([3, 5, 7, 9])
y = np.array([2, 4, 6, 8])
plt.plot(x, y, 'D:r')
plt.show()
```

Marker Size

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([3, 5, 7, 9])
y = np.array([6, 10, 14, 18])
plt.plot(x, y, 'D:r', ms = 20)
plt.show()
```

Marker Color

```python
import matplotlib.pyplot as plt
import numpy as np
y = np.array([16, 10, 14, 9])
plt.plot(y, 'o:r', ms = 20, mec = 'r')
plt.show()
```

Set the FACE color to red

```python
import matplotlib.pyplot as plt
import numpy as np
y = np.array([16, 10, 14, 9])
plt.plot(y, 'o:r', ms = 20, mec = 'r', mfc = 'g')
plt.show()
```
