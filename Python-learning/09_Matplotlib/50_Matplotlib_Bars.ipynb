Creating 4 bars

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array(["A", "B", "C", "D"])
y = np.array([8, 3, 10, 4])
plt.bar(x, y)
plt.show()
```

Another Example practice

```python
import matplotlib.pyplot as plt
import numpy as np

x = ["APPLES", "BANANAS"]
y = [400, 350]

plt.bar(x, y)
plt.show()
```

Horizontal Bars

```python
import matplotlib.pyplot as plt
import numpy as np

x = ["APPLES", "BANANAS"]
y = [400, 350]

plt.barh(x, y)
plt.show()
```

Bar Color

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array(["A", "B", "C", "D"])
y = np.array([8, 3, 10, 4])
plt.bar(x, y, color = 'red')
plt.show()
```

Bar Width

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array(["A", "B", "C", "D"])
y = np.array([8, 3, 10, 4])
plt.bar(x, y, color = 'red', width = 0.2)
plt.show()
```

Bar Height

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array(["A", "B", "C", "D"])
y = np.array([8, 3, 10, 4])
plt.barh(x, y, color = 'red', height = 0.1)
plt.show()
```
