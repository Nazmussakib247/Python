Creating Pie Charts

```python
import matplotlib.pyplot as plt
import numpy as np
y = np.array([4, 6, 8, 10])
plt.pie(y)
plt.show()
```

Add Lables

```python
import matplotlib.pyplot as plt
import numpy as np
y = np.array([4, 8, 12, 16])
mylables = ["Math", "Physcis", "Chemistry", "Biology"]
plt.pie(y, labels =  mylables)
plt.show()
```

Start Angle

```python
import matplotlib.pyplot as plt
import numpy as np
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
plt.pie(y, labels = mylabels, startangle = 90)
plt.show()
```

Explode

```python
import matplotlib.pyplot as plt
import numpy as np
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]
plt.pie(y, labels = mylabels, startangle = 90, explode = myexplode)
plt.show()
```

Shadow

```python
import matplotlib.pyplot as plt
import numpy as np
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]
plt.pie(y, labels = mylabels, startangle = 90, explode = myexplode, shadow = True)
plt.show()
```

Colors

```python
import matplotlib.pyplot as plt
import numpy as np
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]
mycolor = ["Black", "Pink", "Blue", "Green"]
plt.pie(y, labels = mylabels, startangle = 90, explode = myexplode, shadow = True, colors = mycolor)
plt.show()
```

Legend

```python
import matplotlib.pyplot as plt
import numpy as np
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]
plt.pie(y, labels = mylabels, startangle = 90, explode = myexplode, shadow = True)
plt.legend()
plt.show()
```

Legend With Header

```python
import matplotlib.pyplot as plt
import numpy as np
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]
plt.pie(y, labels = mylabels, startangle = 90, explode = myexplode, shadow = True)
plt.legend(title = "Four Fruits:")
plt.show()
```
